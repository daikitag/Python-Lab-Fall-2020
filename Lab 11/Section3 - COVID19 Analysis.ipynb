````python
df
````

````python
from statsmodels.regression.linear_model import OLS
import statsmodels.api as sm

x_vals = df['Date1'].values
y_vals = df['New Cases']

reg_model = OLS(y_vals, sm.add_constant(x_vals)).fit()
display(reg_model.summary())
````

````python
b0 = reg_model.params[0]
b1 = reg_model.params[1]
x_plot = np.linspace(np.min(df['Date1']), np.max(df['Date1']), 100)
````

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.plot(df['Date'], df['New Cases'])
axs.plot(x_plot, x_plot*b1 + b0, color='red')
plt.title("Male Height vs Weight", fontsize=20)
axs.set_xlabel("Weight [kg]", fontsize=18)
axs.set_ylabel("Height [cm]", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

While we can get a brief understanding of the trend, linear relationship might not be the best option to model this. Here, we can have a polynomial regression, where we can model the data to be:
$$
y=\beta_0+\beta_1x+\beta_2x^2+\beta_3x^3+\dots
$$

This is a case of multiple linear regression, where we add multiple predictor variables. You can simply do multiple linear regression in Python by:
```Python
from statsmodels.regression.linear_model import OLS
import statsmodels.api as sm

x_vals = df[['column1','column2']]
y_vals = df['y-value']

reg_model = OLS(y_vals, sm.add_constant(x_vals)).fit()
display(reg_model.summary())
```

````python
df['Date^2'] = df['Date1'] **2
df['Date^3'] = df['Date1'] **3
````

````python
from statsmodels.regression.linear_model import OLS
import statsmodels.api as sm

x_vals = df[['Date1','Date^2', 'Date^3']]
y_vals = df['New Cases']

reg_model = OLS(y_vals, sm.add_constant(x_vals)).fit()
display(reg_model.summary())
````

In regression analysis, we often use R-squared to measure how fit the regression line is. In multiple regression analysis, we often refer to the adjusted R-squared to see the fit of the model. We see a drastic increase in the adjusted R-squared, which indicates that the polynomial regression model fits well to the data.

````python
b0 = reg_model.params[0]
b1 = reg_model.params[1]
b2 = reg_model.params[2]
b3 = reg_model.params[3]
x_plot = np.linspace(np.min(df['Date1']), np.max(df['Date1']), 100)
````

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.plot(df['Date'], df['New Cases'])
axs.plot(x_plot,(x_plot**3)*b3+ (x_plot**2)*b2 + x_plot*b1 + b0, color='red')
plt.title("Male Height vs Weight", fontsize=20)
axs.set_xlabel("Weight [kg]", fontsize=18)
axs.set_ylabel("Height [cm]", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

We now have a much greater fit to the model.

By using this polynomial regression, we can have a much greater prediction of what happens in the future.

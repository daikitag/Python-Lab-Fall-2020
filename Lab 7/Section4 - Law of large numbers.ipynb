# Law of large numbers

Law of large numbers is one of the most important fundamental theorems in statistics that you should know.

````python
import matplotlib.pyplot as plt
import random
````

## Definition:

Formal definition of strong law of large numbers:

- The sample average converges almost surely to the expected value

$$
\bar{X_n}\rightarrow \mu ~~ \text{when} ~~n\rightarrow \infty
$$

$$
P\left(\lim_{n\rightarrow\infty}\bar{X_n}=\mu \right)=1
$$

It means that if I take a sample mean from a really large sample size, it will be extremely close to the actual population mean.

As a first step, we will create a function that computes the probability of heads from the specified number of tosses. Then, we will visualize the results.

````python
# This code is taken from before:
def coin_toss(s):
    result = []
    for i in range(s):
        toss = random.choice(['H','T'])
        result.append(toss)
    return result
````

````python
def head_prob(s):
    toss = coin_toss(s)
    head = toss.count('H')
    prob = head / s
    return prob
````

````python
head_prob(10)
````

````python
head_prob(15)
````

````python
x = []
y = []

for i in range(100):
    y.append(i)
    prob = head_prob(i)
    x.append(prob)
````

I get an error. This is because `i` start from 0, and we can never divide a number by 0. We can rewrite this code as following to let the for loop start from 1:

````python
x = []
y = []

for i in range(1, 101):
    x.append(i)
    prob = head_prob(i)
    y.append(prob)
````

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.plot(x, y)
plt.title('Coin toss simulation', fontsize=20)
axs.set_xlabel('Number of coin tosses', fontsize=18)
axs.set_ylabel('Probability of getting heads', fontsize=18)
axs.axhline(0.5, color='r', ls='--')
axs.tick_params(labelsize=15)
plt.show()
````

As you see, the probability of getting heads will converge to 1/2 as we increase the number of coin tosses.

## 2. Roll a die

````python
def roll_die(n):
    result = []
    for i in range(n):
        die = random.randint(1,6)
        result.append(die)
    die1 = result.count(1) / n
    die2 = result.count(2) / n
    die3 = result.count(3) / n
    die4 = result.count(4) / n
    die5 = result.count(5) / n
    die6 = result.count(6) / n
    prob = [die1, die2, die3, die4, die5, die6]
    return prob
````

Visualize the result

Here, I will be creating multiple plots by specifying integers in
```Python
plt.subplots(a,b)
```

1. 1 Roll

````python
label = [1,2,3,4,5,6]

fig, axs = plt.subplots(3,3,figsize=(15,6))
fig.subplots_adjust(hspace=.5)
for i in range(3):
    for j in range(3):
        axs[i,j].bar(label, roll_die(1), width=0.5)
        axs[i,j].set_ylim(0,1.1)
        axs[i,j].set_title('1 Roll')
````

2. 100 rolls

````python
label = [1,2,3,4,5,6]

fig, axs = plt.subplots(3,3,figsize=(15,6))
fig.subplots_adjust(hspace=.5)
for i in range(3):
    for j in range(3):
        axs[i,j].bar(label, roll_die(100), width=0.5)
        axs[i,j].set_ylim(0,1.1)
        axs[i,j].axhline(1/6, color='r', ls='--')
        axs[i,j].set_title('100 Rolls')
````

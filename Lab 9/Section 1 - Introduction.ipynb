## Hints to HW7 Problem 1

This problem asked you to simulate 100,000 points from Poisson(2) distribution, and approximate the pmf of Poisson(2) through those simulated points. Probability mass function (pmf) is a function that gives the probability that a discrete random variable is exactly equal to some value. Some people may have looked through some online resources, and figured out that the pmf of Poisson distribution can be written as:

$$
f(k;\lambda)=\frac{\lambda^ke^{-\lambda}}{k!}
$$
where $k$ represents the outcome of the Poisson random variable. Let's plot this.

````python
import numpy as np

def pois2(k):
    num = 2**k *np.exp(-2) # 2**k is the same as 2^k in Python
    denom = np.math.factorial(k)
    frac = num / denom
    return frac
````

````python
prob = []
for i in range(11):
    prob.append(pois2(i))
print(prob)
````

Remember the following code to produce the line graph:

```Python

axs.plot(['list of x-values'], ['list of y-values'])

```

````python
import matplotlib.pyplot as plt

x_list = []
for i in range(11):
    x_list.append(i)

fig, axs = plt.subplots(figsize=(12,8))
axs.plot(x_list, prob)
plt.title('Poisson(2) pmf', fontsize=20)
axs.set_xlabel('x values', fontsize=18)
axs.set_ylabel('Probability', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

This is the exact pmf of Poisson(2) distribution. If your plot from the simulated values is close to this, you are in a good shape.

## Brief introduction to confidence intervals

In statistics, we always want to estimate the true parameter from the sample (such as the population mean), but it is usually not possible to determine the exact value of the parameters. Thus, we should solve an inference problem, where we want to estimate the parameter based on the sample statistic:

<img src='Image/slide1.png' width=500>

<img src='Image/slide2.png' width=500>

We usually estimate the population mean within in a range of values. This range is called to be a confidence interval

<img src='Image/slide3.png' width=500>

By using central limit theorem, we can show that the sample statistic (such as the sample average) is expected to be within 2 times standard error of the population parameter with 95\% confidence.

Definition:
- Standard error: standard deviation of the statistics

<img src='Image/slide4.png' width=500>

We are often interested in finding out the 95\% confidence intervals. For sufficiently large sample size, we will multiply the standard error by 1.96 to determine the confidence interval

<img src='Image/slide5.png' width=500>

In summary, you only have to determine the standard error of the sample statistic. That will determine the confidence intervals.

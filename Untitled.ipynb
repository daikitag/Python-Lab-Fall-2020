```python
import pandas as pd
df = pd.read_csv('owid-covid-data.csv')
```

```python
df['continent'].value_counts()
```

```python
asia_df = df[df['continent'] == 'Asia']
europe_df = df[df['continent'] == 'Europe']
```

```python
asia_df
```

```python
asia_df = asia_df[['date', 'total_cases']]
europe_df = europe_df[['date','total_cases']]
```

```python
# I will only create two dataframes. In your answer, you must include all continents!!
```

```python
asia_df
```

```python
date = ['2020-01-01','2020-02-01','2020-03-01','2020-04-01','2020-05-01']
```

```python
import numpy as np

asia_cases = []
for element in date:
    df_new = asia_df[asia_df['date'] == element]
    case = np.sum(df_new['total_cases'])
    asia_cases.append(case)
```

```python
asia_cases
```

```python
europe_cases = []
for element in date:
    df_new = europe_df[europe_df['date'] == element]
    case = np.sum(df_new['total_cases'])
    europe_cases.append(case)
```

```python
europe_cases
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = date
y1 = asia_cases
y2 = europe_cases

y = np.vstack([y1, y2])

labels = ["Asia", "Europe"]

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, labels=labels)
ax.legend(loc='upper left')
plt.show()
```

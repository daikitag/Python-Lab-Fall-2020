# Visualizing BMI values among different race groups

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/NHANES.csv')
```

```python
df = df.dropna(subset=['BMI'])
```

```python
white = df['Race3'] == 'White'
black = df['Race3'] == 'Black'
hispanic = df['Race3'] == 'Hispanic'
asian = df['Race3'] == 'Asian'
```

```python
white_df = df[white]
black_df = df[black]
hispanic_df = df[hispanic]
asian_df = df[asian]
```

## Create grouped histogram

As a first step, we must specify the bins:

```python
print("Max:", max(df['BMI']))
print("Min:", min(df['BMI']))
```

Let's suppose that we want the bin width to be 2. Then,

$$
(82-10)/2 = 72/2 = 36
$$

so we want $(36+1)=37$ points between 10 and 82. We use: `np.linspace(10,82,37)` as bins

```python
fig, axs = plt.subplots(figsize=(12,8))
axs.hist(white_df['BMI'], color="b", alpha=0.3, bins=np.linspace(10,82,37), label="White")
axs.hist(black_df['BMI'], color="r", alpha=0.3, bins=np.linspace(10,82,37), label="Black")
axs.hist(hispanic_df['BMI'], color="y", alpha=0.5, bins=np.linspace(10,82,37), label="Hispanic")
axs.hist(asian_df['BMI'], color="c", alpha=0.5, bins=np.linspace(10,82,37), label="Asian")

plt.title("BMI Categorized by Race", fontsize=20)
axs.set_xlabel("BMI", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
axs.legend(prop={'size': 15})

plt.show()
```

Probably, grouped histogram was not a good choice. Since most of the people are whites, it's extremely difficult to compare the distribution. As a next step, we will create a grouped boxplot.

```python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot([white_df['BMI'], black_df['BMI'], hispanic_df['BMI'], asian_df['BMI']])
plt.title('BMI Categorized by Race', fontsize=20)
axs.set_xticklabels(['White', 'Black', 'Hispanic', 'Asian'])
axs.set_ylabel('BMI', fontsize=18)
axs.tick_params(labelsize=15)
axs.axhline(y=30, color='r')
plt.show()
```

This BMI = 30 is the obese line. While it seems that Asians are less likely to be obese and Blacks are more likely to be obese, it's really hard to compare the differences. This is where you have to use hypothesis testing, which we will cover later. Alternatively, we can visualize the percentage of obese people, which we had done in the previous lab.

We also know that high blood pressure and diabetes are related to COVID-19 cases. Let's examine the relationship between BMI and diabetes.

`Diabetes` column in NHANES dataset represents whether a person has diabetes or not. At first, let's count the values in `Diabetes` column.

```python
df['Diabetes'].value_counts()
```

Most people are healthy, so it would be better to create a comparative box plot.

```python
diabetes = df['Diabetes'] == 'Yes'
healthy = df['Diabetes'] == 'No'
```

```python
diabetes_df = df[diabetes]
healthy_df = df[healthy]
```

```python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot([diabetes_df['BMI'], healthy_df['BMI']])
plt.title('BMI vs Diabetes', fontsize=20)
axs.set_xticklabels(['Diabetes','Healthy'])
axs.set_ylabel('BMI', fontsize=18)
axs.tick_params(labelsize=15)
axs.axhline(y=30, color='r')
plt.show()
```

From this plot, it is obvious that people who have diabetes are more likely to be obese compared to healthy people. Let's create a bar chart that visualizes the percentage differences.

```python
def function(row):
    if row['BMI'] >= 30:
        return 'Obesity'
    elif row['BMI'] < 30:
        return 'Healthy'
    else:
        return 'NaN'
```

```python
diabetes_df['Obesity'] = diabetes_df.apply(function, axis=1)
healthy_df['Obesity'] = healthy_df.apply(function, axis=1)
```

```python
table = pd.crosstab(index=df["Race3"], columns=df["Diabetes"]) 
table
```

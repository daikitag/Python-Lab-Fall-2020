In this notebook, we will present some useful techniques that you could use to make your histograms look better

## Import data science trio

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

We always have to import these three packages in the beginning. You only have to import them once, so it would be useful to import them at first.

Data visualization is like an art, and there are many ways to improve the visual. We will be manipulating this histogram that we produced in the previous section:

````python
df = pd.read_csv('Data/NHANES.csv')
height = df['Height'].dropna()

x_value = height

fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black", bins=np.linspace(80,202,62))
plt.title("Height", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

## 1. Add vertical line

Vertical lines can be added by using the following function:

```Python
axs.axvline(x, color='red', ls='dashed')
```
`x` specifies the x-coordinate, `color` can change the color, and `ls` changes the linestyle. `'dashed'` and `'solid'` are the main options that are commonly used

Reference:

https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.axes.Axes.axvline.html

We will be using this to show mean and median, as they are not easy to understand from just observing the histogram.

````python
x_value = height

fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black", bins=np.linspace(80,202,62))
plt.title("Height", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
# I added two lines here:
axs.axvline(np.mean(height), color='red', ls='dashed')
axs.axvline(np.median(height), color='blue', ls='dashed')

plt.show()
````

We can also include `lw` argument in `axs.axvline()` to make the line width to be larger:

````python
x_value = height

fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black", bins=np.linspace(80,202,62))
plt.title("Height", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
# I added two lines here:
axs.axvline(np.mean(height), color='red', ls='dashed', lw=3)
axs.axvline(np.median(height), color='blue', ls='dashed', lw=3)

plt.show()
````

It looks better now. As a next step, we will add a legend to the plot indicating what the lines represent.

## 2. Legend

If you would like to label the object in a plot, you have to put `label=` argument in the code. Here, change 
```Python
axs.axvline(x, color='red', ls='dashed')
```
to be:
```Python
axs.axvline(x, color='red', ls='dashed', label='label')
```
and change the `label` argument.

At the last part of the code, add:
```Python
axs.legend(prop={'size': 15})
```
You can adjust the number to change the size of the legend.

````python
x_value = height

fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black", bins=np.linspace(80,202,62))
plt.title("Distribution of height in the U.S.", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
# I added two lines here:
axs.axvline(np.mean(height), color='red', ls='dashed', lw=3, label='Mean')
axs.axvline(np.median(height), color='blue', ls='dashed', lw=3, label='Median')
axs.legend(prop={'size': 15})

plt.show()
````

Adding a legend was simple. Compare this with the original plot. This plot looks better.

````python
x_value=height
fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black")
plt.title("Height", fontsize=20)
axs.set_xlabel("Height[cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

# Section4 - Grouped Histograms

We are now interested in created grouped histograms. Remember the previous lab that we separated the original dataframe (height data) based on gender and age. We will do the same things here at first and then create a grouped histogram.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('Data/NHANES.csv')
df = df[['Gender', 'Age', 'Height']]
df = df.dropna()
````

## Separate them by condition

````python
adult = df['Age'] >= 20
child = df['Age'] < 20

# Remember that == means a condition statement to find if two elements are the same
male = df['Gender'] == 'male'
female = df['Gender'] == 'female'

# We will create two separate dataframes for male and female / adult and child

male_df = df[adult & male]
female_df = df[adult & female]
child_df = df[child]
````

Now that we have separate dataframe for adult/child and male/female, we will be using the `'Height'` column in these dataframes. Grouped bar chart can be produced by just adding many `axs.hist()` functions in the plot.

Remember that the histogram for 1 variable looked like:

```Python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(12,8))
axs.hist(x_value, color="SkyBlue", edgecolor="Black")
plt.title("Title", fontsize=20)
axs.set_xlabel("x element", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

We add `axs.hist()` while adjusting the color, and the bins have to be the same in all bars.

````python
# modify the codes during classtime
# remind them to type
# You can change things in green and red
````

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.hist(male_df['Height'], color="b", alpha=0.3, bins=np.linspace(80,202,62), label="Male Adult")
axs.hist(female_df['Height'], color="r", alpha=0.3, bins=np.linspace(80,202,62), label="Female Adult")
axs.hist(child_df['Height'], color="y", alpha=0.5, bins=np.linspace(80,202,62), label="Child")

plt.title("Distribution of height in the U.S.", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
axs.legend(prop={'size': 15})

plt.show()
````

We were able to show the histogram of different groups in one plot. It seems that male/female adult heights follow a symmetric distribution. There are few things that you must remember:

1. You must specidy the `bins` to be the same for all groups. Otherwise, the plot will look very awkward
2. Remember that `alpha` can change the darkness of the colors. Since we want them to be overlapping, you should select the alpha to be a small value to see the overlapping part
3. Plot dark colors first. Here, red and green are much darker than yellow, so you should plot them first

It is also possible to add lines to represent means of different groups.

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.hist(male_df['Height'], color="b", alpha=0.3, bins=np.linspace(80,202,62), label="Male Adult")
axs.hist(female_df['Height'], color="r", alpha=0.3, bins=np.linspace(80,202,62), label="Female Adult")
axs.hist(child_df['Height'], color="y", alpha=0.5, bins=np.linspace(80,202,62), label="Child")

plt.title("Distribution of height in the U.S.", fontsize=20)
axs.set_xlabel("Height [cm]", fontsize=18)
axs.set_ylabel("Frequency", fontsize=18)
axs.tick_params(labelsize=15)
axs.legend(prop={'size': 15})

axs.axvline(np.mean(male_df['Height']), color='b', ls='dashed', lw=2)
axs.axvline(np.mean(female_df['Height']), color='r', ls='dashed', lw=2)
axs.axvline(np.mean(child_df['Height']), color='y', ls='dashed', lw=2)

plt.show()
````

# Section5 - Boxplot

Creating a boxplot can be very simple. It can be produced by using the following code:

```Python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot(['list of values'])
plt.title('title', fontsize=20)
axs.set_xticklabels(['list of labels'])
axs.set_ylabel('y-axis labels', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

## Importing packages + Data preparation

````python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('Data/NHANES.csv')
df = df[['Gender', 'Age', 'Height']]
df = df.dropna()
````

## 1. Create boxplot with 1 value

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot(df['Height'])
plt.title('Height distribution of the U.S.', fontsize=20)
axs.set_xticklabels([''])
axs.set_ylabel('Height [cm]', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

This shows a negatively skewed distribution. We will be separating them by adult/children and male/female.

## 2. Grouped boxplot

````python
adult = df['Age'] >= 20
child = df['Age'] < 20

# Remember that == means a condition statement to find if two elements are the same
male = df['Gender'] == 'male'
female = df['Gender'] == 'female'

# We will create two separate dataframes for male and female / adult and child

male_df = df[adult & male]
female_df = df[adult & female]
child_df = df[child]
````

````python
male_height = male_df['Height']
female_height = female_df['Height']
child_height = child_df['Height']
````

````python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot([male_height, female_height, child_height])
plt.title('Height distribution of the U.S.', fontsize=20)
axs.set_xticklabels(['Male Adult','Female Adult','Children'])
axs.set_ylabel('Height [cm]', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
````

Now, we can understand the distribution of male adult, female adult and children. For grouped data, boxplots might be a better option than histograms.

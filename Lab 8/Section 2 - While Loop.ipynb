# 2. While loops

While loops are very similar to for loops. The basic syntax is:

```Python
while condition:
    do something
```

````python
i = 0
while i < 4:
    print(i)
    i += 1
````

Here, you MUST write `i += 1` at the end, because while loops don't automatically add 1 to `i` like for loop. So, the following code is incorrect:

````python
i = 0
while i < 4:
    print(i)
````

One of the largest advantages of using while loop is the fact that it never ends unless the condition becomes `False`. However, you should always make sure to write a while loop, so that it ultimately ends.

While loops can start from the biggest number:

````python
i = 5
while i > 1:
    print(i)
    i -= 1
````

You can also use while loops, so that the distance between two numbers is greater than 1.

For example:

Write a program that gives 1, 4, 7, 10 as its output:

````python
i = 1
while i <= 10:
    print(i)
    i += 3
````

While loop to find if the number is prime number or not:

````python
number = 19

i = 2
while i < number:
    if number % i == 0:
        print(number, 'is not a prime number')
        break
    elif i == number -1:
        print(number, 'is a prime number')
    i += 1
````

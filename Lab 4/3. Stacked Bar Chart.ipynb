As a next step, we will create a stacked bar chart from this dataframe. You can create a stacked bar chart by adding `axs.bar()` to the code that we created a bar chart.

```Python
import matplotlib.pyplot as plt

label = []
y_value1 = []
y_value2 = []
y_value3 = []

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
p1 = axs.bar(label, y_value1, color='blue')     
# Don't forget the bottom from p2
p2 = axs.bar(label, y_value2, color='red', bottom=y_value1)
p3 = axs.bar(label, y_value3, color='green', bottom=y_value1+y_value2)
axs.set_title("Plot Title", fontsize=20, fontweight="bold")   
axs.set_xlabel("x-axis label", fontsize=14)
axs.set_ylabel("y-axis label", fontsize=14)
axs.tick_params(labelsize=16)  
axs.legend((p1[0],p2[0],p3[0]),('label1','label2','label3'),
          bbox_to_anchor=(1.05, 1))
# make sure to add p here, if you would like to add more bars
plt.show() 
```
`bottom` specifies the y-axis location to start the bar chart, and `bbox_to_anchor=(1.05, 1)` specifies that you would like to have the legend outside the plot. You can delete it to put the legend inside the plot.

````python
import matplotlib.pyplot as plt

label = ['Asian', 'Black', 'Hispanic', 'Mexican', 'Other', 'White']
y_value1 = table.loc['Underweight']
y_value2 = table.loc['Normal']
y_value3 = table.loc['Overweight']
y_value4 = table.loc['Obesity']
y_value5 = table.loc['Severe Obesity']

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here    

p1 = axs.bar(label, y_value1) # You specify the color here     
p2 = axs.bar(label, y_value2, bottom=y_value1)
p3 = axs.bar(label, y_value3, bottom=(y_value1+y_value2))
p4 = axs.bar(label, y_value4, bottom=(y_value1+y_value2+y_value3))
p5 = axs.bar(label, y_value5, bottom=(y_value1+y_value2+y_value3+y_value4))

axs.set_title("Obesity level broken down by race", fontsize=20, fontweight="bold")   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("Proportion", fontsize=14)
axs.tick_params(labelsize=16)  
axs.legend((p1[0],p2[0],p3[0],p4[0],p5[0]),(table.index), bbox_to_anchor=(1.05, 1))
# make sure to add p here, if you would like to add more bars
plt.show() 
````

This looks far better compared to last time. Sometimes, we would like to compare the obesity levels excluding 'underweight', 'normal' and 'overweight' to compare the obese percentage.

````python
import matplotlib.pyplot as plt

label = ['Asian', 'Black', 'Hispanic', 'Mexican', 'Other', 'White']
y_value1 = table.loc['Obesity']
y_value2 = table.loc['Severe Obesity']

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here    

p1 = axs.bar(label, y_value1) # You specify the color here     
p2 = axs.bar(label, y_value2, bottom=y_value1)

axs.set_title("Obesity level broken down by race", fontsize=20, fontweight="bold")   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("Percentage[%]", fontsize=14)
axs.tick_params(labelsize=16)  
axs.legend((p1[0],p2[0]),('Obesity','Severe Obesity'))
# make sure to add p here, if you would like to add more bars
plt.show() 
````

Now, we can compare the obese percentages between different race groups, while understanding the proportion of 'severe obesity' and 'obesity' as well.

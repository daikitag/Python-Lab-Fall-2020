# Bar chart

This is the code that can create a bar chart:

```Python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
axs.bar(x_values, y_values, color=('deepskyblue')) # You specify the color here     
axs.set_title("Plot Title", fontsize=20, fontweight="bold")   
axs.set_xlabel("x-axis label", fontsize=14)
axs.set_ylabel("y-axis label", fontsize=14)
axs.tick_params(labelsize=16)              
plt.show() 
# You can change the fontsize by changing the numbers after fontsize=
```

## Add numerical labels to bar chart

You can use the following code to add numerical labels to bar chart

```Python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
bargraph = axs.bar(x_values, y_values, color=('blue'), alpha=0.4) # You specify the color here     
axs.set_title("Plot Title", fontsize=20, fontweight="bold")   
axs.set_xlabel("x-axis label", fontsize=14)
axs.set_ylabel("y-axis label", fontsize=14)
# axs.set_ylim(0,100)  # It's better to set this for better visualization purposes
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

for i in range(len(x_values)):
    plt.text(i, y_values[i], str(y_values[i]), color='b', size=16)

plt.grid(axis='y')

plt.show() 
```

## 2. Copy and paste the code that is given

We will be plotting "Deaths".

````python
x_values = df['Race'] # This specifies the label
y_values = df['Deaths'] # This specifies the value

import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
bargraph = axs.bar(x_values, y_values, color=('blue'), alpha=0.4) # You specify the color here     
axs.set_title("Plot Title", fontsize=20, fontweight="bold")   
axs.set_xlabel("x-axis label", fontsize=14)
axs.set_ylabel("y-axis label", fontsize=14)
# axs.set_ylim(0,100)  # It's better to set this for better visualization purposes
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

for i in range(len(x_values)):
    plt.text(i, y_values[i], str(y_values[i]), color='b', size=16)

plt.grid(axis='y')

plt.show()
````

````python
x_values = df['Race'] # This specifies the label
y_values = df['Deaths'] # This specifies the value

import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
bargraph = axs.bar(x_values, y_values, color=('blue'), alpha=0.4) # You specify the color here     
axs.set_title("COVID-19 Deaths Broken Down by Race", fontsize=20, fontweight="bold")   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("COVID-19 Deaths", fontsize=14)
# axs.set_ylim(0,100)  # It's better to set this for better visualization purposes
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

for i in range(len(x_values)):
    plt.text(i, y_values[i], str(y_values[i]), color='b', size=16)

plt.grid(axis='y')

plt.show()
````

Changing the location of these labels require some trial and error.
```Python
plt.text(i, y_values[i], str(y_values[i]), color='b', size=16)
```
plots text in `(i, y_values[i])` coordinate. Here, notice that you have to shift the text to left (negative x-value) and upward (positive y-value).

The values in x-axis are integers (1,2,...), so you only have to shift it slightly. However, the values in the y-axis here are 1000,000 so you will have to shift it greater than x value.

## 1. Shift -0.3 x and + 10,000 y

````python
x_values = df['Race'] # This specifies the label
y_values = df['Deaths'] # This specifies the value

import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
bargraph = axs.bar(x_values, y_values, color=('blue'), alpha=0.4) # You specify the color here     
axs.set_title("COVID-19 Deaths Broken Down by Race", fontsize=20, fontweight="bold")   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("COVID-19 Deaths", fontsize=14)
# axs.set_ylim(0,100)  # It's better to set this for better visualization purposes
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

for i in range(len(x_values)):
    plt.text(i-0.3, y_values[i]+10000, str(y_values[i]), color='b', size=16) # I changed the values here

plt.grid(axis='y')

plt.show()
````

Now, the x-axis looks totally fine, but the texts have to move downward. We will next try

Looks great! However, notice that the space between the title and the plot is too small. We can set the axis limit in

by using this code. The numbers inside `()` represent the range of the limit. The range of the original plot was (0,100000) by default (see the y-axis values), so we will slightly increase the y-limit.

## 3. Set y-limit to be (0, 110000)

````python
x_values = df['Race'] # This specifies the label
y_values = df['Deaths'] # This specifies the value

import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10,6)) # Change the figure size here       
bargraph = axs.bar(x_values, y_values, color=('blue'), alpha=0.4) # You specify the color here     
axs.set_title("COVID-19 Deaths Broken Down by Race", fontsize=20, fontweight="bold")   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("COVID-19 Deaths", fontsize=14)
axs.set_ylim(0,110000)  # It's better to set this for better visualization purposes
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

for i in range(len(x_values)):
    plt.text(i-0.3, y_values[i]+2000, str(y_values[i]), color='b', size=16) # I changed the values here

plt.grid(axis='y')

plt.show()
````

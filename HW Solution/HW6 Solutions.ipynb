## Greater money or more luck?

I have two gambers:

- GamblerA: 1,000 dollars at the beginning, and his chance of winning against GamberB is 0.4
- GamblerB: 200 dollars at the beginning, and his chance of winning againt GamberA is 0.6

GamblerA plays a game with GamberB. In each of these games, the gambler wins 100 dollars if he wins the game, and loses 100 dollars otherwise. The game continues until one player has 0 dollars left (the other person must have a total of 1,200 dollars at the end).

By using simulation, determine the probability that GamberB wins against GamblerA.

Grading guideline:
- Create a function that conducts simulation [10 points]
- Use print statement to show the probability that GamblerA wins [3 points]
- Comment on which gambler had the higher possibility of winning [2 points] (Do Not forget to write this)

### Hint:
You can combine if statement and `random.randint()` function to simulate probability = 0.4 and 0.6

```python
import random
```

```python
def gamble(s):
    personA = 1000
    personB = 200
    for i in range(s):
        out = random.randint(1,5)
        if out == 1 or out == 2:
            personA += 100
            personB -= 100
        else:
            personA -= 100
            personB += 100
        if personA == 0:
            return 'B'
            break
        elif personB == 0:
            return 'A'
            break
```

```python
person = []
for i in range(10000):
    win = gamble(1000)
    person.append(win)
```

```python
probA = person.count('A') / 10000
```

```python
print("The probability that GamberA wins is:", probA)
```

Gambler A's probability of winning is approximately 44%. Gambler A's probability of winning 1 game is 40%, so it is higher than that, but simulation shows that it would be better to have a higher chance of winning 1 game, instead of having a large amount of money at the beginning.
